```python
print("Rows, Cols:", df.shape)
```

```python
print("disaster_type counts:")
print(df["disaster_type"].value_counts(dropna=False))


required = {"none", "flood", "fire", "earthquake"}
present = set(df["disaster_type"].dropna().unique())
missing = required - present
print("Missing disaster types:", missing if missing else "None")

```

```python
print("blocked_ratio min/max:", df["blocked_ratio"].min(), df["blocked_ratio"].max())
print("blocked_ratio > 0 rows:", (df["blocked_ratio"] > 0).sum())
```

```python
print("evac_rate min/max:", df["evac_rate"].min(), df["evac_rate"].max())
print("evac_rate unique values:", df["evac_rate"].nunique())
print("evac_rate std:", df["evac_rate"].std())
```
